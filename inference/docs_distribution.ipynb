```python
def print_intersection_counts(intersection_df):
    """교집합 개수별 ID 개수 출력"""
    counts = intersection_df['all_intersection'].value_counts().sort_index()
    
    print(f"\n총 ID 개수: {len(intersection_df)}")
    print("\n=== 교집합 개수별 ID 개수 ===")
    for intersection_count, id_count in counts.items():
        print(f"교집합 {intersection_count}개: {id_count}개 ID")
```
